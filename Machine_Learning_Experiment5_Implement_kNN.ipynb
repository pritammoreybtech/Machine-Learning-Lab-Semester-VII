```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import pandas as pd

#Small sample dataset (2D points)
#Features: [height (cm), weight(kg)]

X = [
    [150, 50],
    [160, 60],
    [170, 70],
    [180, 80],
    [190, 85],
    [200, 90]
]

#Labels: 0 = Class A, 1 = Class B
y = [0, 0, 0, 1, 1, 1]

#Create and train the model
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)

#Test on new points
X_test = [
    [155, 55],  #likely Class A
    [185, 82] #likely Class B
]
y_test = [0,1]

#Predict
y_pred = knn.predict(X_test)
print("Predictions: ", y_pred)

#Evaluate
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Class A", "Class B"])
disp.plot()
plt.title("Confusion Matrix")
plt.show()
```
